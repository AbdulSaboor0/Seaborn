```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

### NOTE : if we are using this to load the dataset the type will not be a dataframe
tips_data = sns.load_dataset("tips")
tips_data.head(5)
```

```python
tips_data.info()
```

```python
tips_data.describe()
```

```python
# Apply the background theme
sns.set_theme()

sns.scatterplot(x='total_bill', y = 'tip', data = tips_data)


### We can also set the Titles by matplot
plt.title("Scatterplot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
```

```python

sns.lineplot(x = 'size', y = 'total_bill', data = tips_data)
```

```python
### Now for Categorical values we have Categorical plots

### 1. Bar Plot
sns.barplot(x = "day", y ="total_bill", data = tips_data)
plt.title("Day VS Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

```python
### 2. Box Plot
sns.boxplot(x = "day", y = "total_bill", data = tips_data)
plt.title("Day VS Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

```python
### To check the distribution of the data
### 3. Violin Plot

sns.violinplot(x = "day", y ="total_bill", data =tips_data)
```

```python
### Histograms
sns.set_style("whitegrid"
)
sns.histplot(tips_data['total_bill'], bins = 7, kde = True)
```

```python
### KDE PLOT

sns.kdeplot(tips_data["total_bill"], fill = True, color = "g")
```

```python
### Pair Plot (Numerical Data only)

sns.pairplot(tips_data)
```

```python
Corr = tips_data[["size","tip","total_bill"]].corr()
Corr
```

```python
#### Plot color gradients['viridis', 'plasma', 'inferno', 'magma', 'cividis'])
#### Plot colors['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds','YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu','GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])


sns.heatmap(Corr, annot = True, cmap= "Blues")
```
